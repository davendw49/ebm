```python
def cal_baseline_upper_lower(C):
    r = 1
    alpha_lower = (1+C)/r
    alpha_upper = C+r
    return alpha_lower, alpha_upper
```

```python
def cal_real_upper_lower(C,r=5):
    alpha_lower = (1+C)/r
    alpha_upper = C+r
    return alpha_lower, alpha_upper
```
