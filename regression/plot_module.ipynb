```python
alist = [(355234.3394,355083.9463),(1013096.972,1012500.646),(159947.8862,159835.0362),  
         (89724.90926,89624.46625),(5582.597239,5556.626213),(2854000.851,2853724.53),
         (8145.4605,8079.2388),(19143.6870,19066.3798),(22305.2852,22214.9528),(55145.696662021866,55093.48075222508)] #(92.8342,91.4423)
namelist = ['Book\nCorssing','Audio\nScrobbler','Amazon\nCD','Amazon\nMovies','Amazon\nElectronics','Amazon\nBook','Algorithm','Programming\nLanguage','Artificial\nIntelligence','Libimseti.cz'] #'Facebook'
```

```python
def show_dist(param, i, name):
    # print(i)
    plt.subplot(2,5,i+1)
    a = param[0]
    b = param[1]
    num_list1 = [a]
    num_list2 = [b]
    x = range(len(num_list1))
    rects1 = plt.bar(x, width=0.1, height=num_list1, color='dimgray')
    rects2 = plt.bar([i + 0.1 for i in x], width=0.1, height=num_list2, color='silver')
    plt.ylim(a-5*(abs(a-b)), b+2*(abs(a-b)))
    plt.xticks([])
    plt.yticks(rotation=45)
    # 编辑文本
#     for rect in rects1:
#         height = rect.get_height()
#         plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height), ha="center", va="bottom",rotation=-45)
#     for rect in rects2:
#         height = rect.get_height()
#         plt.text(rect.get_x() + rect.get_width() / 2, height+1, str(height), ha="center", va="bottom",rotation=-45)
    # plt.savefig("weight.png",bbox_inches = 'tight')
    ax = plt.gca()      #gca='get current axis'
    # 将右边和上边的边框（脊）的颜色去掉
    ax.spines['right'].set_color('none')
    ax.spines['top'].set_color('none')
    ax = plt.gca()
    plt.title(name, loc='center', fontsize=14)
    # plt.show()
```

```python
if __name__=="__main__":
    biggraph = plt.figure(figsize=(10, 5))
    for i in range(10):
        show_dist(alist[i], i, namelist[i])
    biggraph.legend(['Baseline', 'Mixture case'], loc='lower center',ncol=2, fontsize=16)
    plt.subplots_adjust(wspace=0.5, hspace =0.4)
    plt.savefig("regression.eps",bbox_inches = 'tight',dpi=600)
    plt.savefig("regression.png",bbox_inches = 'tight',dpi=600)
    plt.show()
```
